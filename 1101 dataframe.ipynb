```python
#刪除觀測值或欄位
import pandas as pd

groups = ["one", "two", "three", "four", "five", "six"]
ironmen = [59, 9, 19, 14, 6, 77]

ironmen_dict = {
                "groups": groups,
                "ironmen": ironmen
}

# 建立 data frame
ironmen_df = pd.DataFrame(ironmen_dict)
print(ironmen_df)


#刪除觀測值
ironmen_df_no_mw = ironmen_df.drop(0,axis=0)
print(ironmen_df_no_mw)

ironmen_df_no_mw = ironmen_df.drop(1,axis=0)
print(ironmen_df_no_mw)

#刪除欄位
ironmen_df_no_groups = ironmen_df.drop("groups",axis=1)
print(ironmen_df_no_groups)

```

```python
#透過 ix 屬性篩選 data frame

import pandas as pd

groups = ["one", "two", "three", "four", "five", "six"]
ironmen = [59, 9, 19, 14, 6, 77]

ironmen_dict = {
                "groups": groups,
                "ironmen": ironmen
}

# 建立 data frame
ironmen_df = pd.DataFrame(ironmen_dict)
print(ironmen_df)
print("---------------")
#選擇欄位
print( ironmen_df.ix[:,"groups"])
print("---------------")

#選擇觀測值
print(ironmen_df.ix[0])
print("---------------")

#同時選擇
print( ironmen_df.ix[0,"groups"])
print("---------------")
print( ironmen_df.ix[0,0])
```

```python
#排序
#sort_index() 方法

import pandas as pd

groups = ["3", "2", "6", "5", "1", "4"]
ironmen = [59, 9, 19, 14, 6, 77]

# 用index建立 data frame
ironmen_df = pd.DataFrame(ironmen, columns = ["ironmen"], 
                          index = groups)

print(ironmen_df)
print("---------------")

#用索引值排序
ironmen_df.sort_index()
```

```python
#排序
#sort_values() 方法

import pandas as pd

groups = ["3", "2", "1", "5", "6", "4"]
ironmen = [59, 9, 19, 14, 6, 77]

# 用index建立 data frame
ironmen_df = pd.DataFrame(ironmen, columns = ["ironmen"], 
                          index = groups)

print(ironmen_df)
print("---------------")

#用數值排序
ironmen_df.sort_values(by="ironmen")
```

```python
#描述統計
# sum()、mean()、median() 與 describe() 
import pandas as pd

groups = ["one", "two", "three", "four", "five", "six"]
ironmen = [59, 9, 19, 14, 6, 77]

ironmen_dict = {
                "groups": groups,
                "ironmen": ironmen
}

# 建立 data frame
ironmen_df = pd.DataFrame(ironmen_dict)
print(ironmen_df)
print("---------------")

#sum()
print( ironmen_df.sum())
print("---------------")

#mean()
print( ironmen_df.mean())
print("---------------")

#median
print( ironmen_df.median())
print("---------------")

#describe
print( ironmen_df.describe())
print("---------------")
```

```python
#相異值個數
#value_counts()
import pandas as pd

gender = ["Male", "Male", "Female", "Male", "Male", "Female",]
ironmen = [59, 59, 19, 14, 6, 77]

# 用index建立 data frame
ironmen_df = pd.DataFrame(gender, columns = ["gender"], 
                          index = ironmen)
print(ironmen_df)
print("---------------")

#計算各有幾個觀測值
pd.value_counts( ironmen_df.gender)
```

```python
#判斷遺失值
#isnull() 方法
#notnull() 方法
import numpy as np
import pandas as pd

groups = ["one", "two", np.nan, "four", "five", "six"]
ironmen = [59, 9, 19, 14, 6, np.nan]

ironmen_dict = {
                "groups": groups,
                "ironmen": ironmen
}

# 建立 data frame
ironmen_df = pd.DataFrame(ironmen_dict)
print(ironmen_df)
print("---------------")

print( ironmen_df.ix[:,"groups"].isnull())
print( ironmen_df.ix[:,"ironmen"].notnull())
```

```python
#處理遺失值
#dropna() 方法
#fillna() 方法
import numpy as np
import pandas as pd

groups = ["one", "two", np.nan, "four", "five", "six"]
ironmen = [59, 9, 19, 14, 6, np.nan]

ironmen_dict = {
                "groups": groups,
                "ironmen": ironmen
}

# 建立 data frame
ironmen_df = pd.DataFrame(ironmen_dict)
print(ironmen_df)
print("---------------")

# 有遺失值的觀測值都刪除
ironmen_delete = ironmen_df.dropna()
print(ironmen_delete)
print("---------------")

# 有遺失值的觀測值填補 0
ironmen_fill = ironmen_df.fillna(0)
print(ironmen_fill)
print("---------------")

# 依欄位填補遺失值
fill = ironmen_df.fillna( {"groups": "number",
                          "ironmen":100})
print(fill)
```
