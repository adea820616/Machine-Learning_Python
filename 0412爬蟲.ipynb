```python
import requests

def get_page(url):
    html_doc = requests.get(
        url = url, #url = "網址"
        cookies = { "over18" : "1" }) #PTT第一頁是點選是否滿18歲，設cookie已滿 "1"是18
    return html_doc.text

print( get_page("https://www.ptt.cc/bbs/Gossiping/index39377.html") )
```

```python
#傳回的文字為list型態
def get_articles( dom ):
    articles = []   
    soup = BeautifulSoup (page, "html.parser" )
    divs = soup.find_all( class_="r-ent")
    for obj in divs:
        #如果文章沒被刪除的話
        a = obj.find( class_="title").find("a")
        if a:
            articles.append({
                "title": a.text,
                "href":a["href"]
            })
    return articles

page = get_page("https://www.ptt.cc/bbs/Gossiping/index39377.html")
#print(get_articles(page))

articles = get_articles( page )
for arti in articles:
    #print( arti["href"])
    page1 = get_page("https://www.ptt.cc/" + arti["href"])
    print(page1)
    
```
