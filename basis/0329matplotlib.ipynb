```python
import matplotlib as mpl
import matplotlib.pyplot as plt
#import seaborn #幫助matplotlib畫的比較漂亮

#"color" or "point
#"bo" "b-" "b^
plt.plot( [1,3,5,7], [2,4,6,20] ,"b^") 

plt.xlabel( "x axis")
plt.ylabel( "y axis")

plt.show()
```

```python
import numpy as np
x = np.arange(0., 10. ,.5)

#plt.plot(x ,x**2, "bo", x, x**3, "gs")
plt.plot(x, np.sin(x), "b-", x, np.cos(x), "g-")

plt.show()
```
